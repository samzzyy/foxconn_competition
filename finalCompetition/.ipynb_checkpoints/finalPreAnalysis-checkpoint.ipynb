```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
from mpl_toolkits.mplot3d import Axes3D
def plt_3d(df_temp):
    fig = plt.figure(figsize=(12,8))
    ax = Axes3D(fig)
    ax.plot(df_temp["x"],df_temp["y"],df_temp["z"],c="b")
    plt.show
def plt_plot(y_real):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(range(len(y_real)),y_real, color='b')
    plt.show()
```

```python
final_dire='../../final/03-FinalData-r1el/'
file_dire=os.path.join(final_dire,'0'+str(1),'Sensor')
file_list=os.listdir(file_dire)
```

```python
for tool_index in range(1,5):
    print("final",tool_index)
    file_dire=os.path.join(final_dire,'0'+str(tool_index),'PLC')
    df_temp=pd.read_csv(os.path.join(file_dire,"plc.csv"))
    plt_3d(df_temp)
    plt_plot(df_temp["spindle_load"])
```
