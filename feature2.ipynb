```python
import pandas as pd
from pandas import DataFrame as df
import numpy as np
import os
from scipy import stats
import gc
import time
```

```python
def get_feature(data):
    # count feature
    tmp = []
#     residual_life = int(lifespan - (inxnum) * 5)
#     tmp.append(residual_life)
    toolage = int((inxnum) * 5)
    tmp.append(toolage)
    print(tmp)
    
    absmean = abs(data).mean()
    absmedia = abs(data).median()
#     absmode = stats.mode(abs(data))[0][0]
#     absmode = pd.Series(absmode)
    absrange = abs(data.max()) - abs(data.min())
    absstdn = abs(data).std()
    absvar = abs(data).var()
    maxvalue = data.max()
    absmax = abs(data).max()
    minvalue = data.min()
    absmin = abs(data).min()
#     result = pd.concat([absmean, absmedia, absmode, absrange, absstdn, absvar, maxvalue, absmax, minvalue, absmin])
    result = pd.concat([absmean, absmedia, absrange, absstdn, absvar, maxvalue, absmax, minvalue, absmin])
    result = result.append(pd.Series(tmp))
    result = df(result)
    return result
```

```python
source_dire='../01-TrainingData-additional'
# source_dire='../02-TestingData-keD1'
```

```python
for dire_index in [1,2,3]:
    root_path=os.path.join(source_dire,'0'+str(dire_index),'Sensor1')
    chunk = 25600
    df_feature = df()
    filelist = os.listdir(root_path)
    length = len(filelist)
    lifespan = length*5
    for inxnum in range(1, length+1):
        print("now is in {} file".format(inxnum))
        filepath = os.path.join(root_path, str(inxnum)+".csv")
        handledata = pd.read_csv(filepath, usecols=['vibration_1', 'vibration_2', 'vibration_3', 'current'])
        rownum = handledata.shape[0]
        retimes = rownum // chunk
        for rt in range(retimes):
            data = handledata[rt*chunk : (rt+1)*chunk]
            feature = get_feature(data)
            df_feature = pd.concat([df_feature, feature],axis=1)
            print(df_feature.shape)
    df_feature = df_feature.T
    df_feature.columns = ["v1_absmean", "v2_absmean","v3_absmean","c_absmean",
                          "v1_absmedia", "v2_absmedia","v3_absmedia","c_absmedia",
#                           "v1_absmode", "v2_absmode", "v3_absmode","c_absmode",
                          "v1_absrange", "v2_absrange","v3_absrange","c_absrange",
                          "v1_absstdn", "v2_absstdn","v3_absstdn","c_absstdn",
                          "v1_absvar",  "v2_absvar", "v3_absvar", "c_absvar",
                          "v1_maxvalue", "v2_maxvalue","v3_maxvalue","c_maxvalue",
                          "v1_absmax", "v2_absmax","v3_absmax","c_absmax",
                          "v1_minvalue", "v2_minvalue","v3_minvalue","c_minvalue",
                          "v1_absmin","v2_absmin", "v3_absmin","c_absmin",
                          "tool_age"]
    print(df_feature.shape)
    df_feature.to_csv("feature/{}_train_additional_feature.csv".format(dire_index), index=False)
    gc.collect()
    del df_feature
```
