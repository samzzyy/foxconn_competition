```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
def plt_plot(x,y):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(x,y, c="b")
    plt.show()
```

```python
data_dire="../02-TestingData-poL3/02/Sensor"
# data_dire="../02-TestingData-poL3/04/PLC"
# data_dire="../01-TrainingData-qLua/01/Sensor"
for file_index in range(1,11):
    df_signal=pd.read_csv(os.path.join(data_dire,str(file_index)+".csv"))
#     print(df_signal.shape,df_signal.columns)
    print(file_index)

    worng_index_list=df_signal[(df_signal['vibration_1']>70)|(df_signal['vibration_1']<-70)].index
    for index in worng_index_list:
        df_signal['vibration_1'][index]=df_signal['vibration_1'][index-1]
    worng_index_list=df_signal[(df_signal['vibration_2']>70)|(df_signal['vibration_2']<-70)].index
    for index in worng_index_list:   
        df_signal['vibration_2'][index]=df_signal['vibration_2'][index-1]
    worng_index_list=df_signal[(df_signal['vibration_3']>70)|(df_signal['vibration_3']<-70)].index
    for index in worng_index_list:   
        df_signal['vibration_3'][index]=df_signal['vibration_3'][index-1]
    worng_index_list=df_signal[(df_signal['current']>70)|(df_signal['current']<-70)].index
    for index in worng_index_list:   
        df_signal['current'][index]=df_signal['current'][index-1]
    plt_plot(range(df_signal.shape[0]),df_signal["vibration_1"])
    plt_plot(range(df_signal.shape[0]),df_signal["vibration_2"])
    plt_plot(range(df_signal.shape[0]),df_signal["vibration_3"])
    plt_plot(range(df_signal.shape[0]),df_signal["current"])
    df_signal.to_csv(os.path.join(data_dire,str(file_index)+".csv"),index=False)
```
