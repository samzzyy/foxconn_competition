```python
import pandas as pd
import os 
import matplotlib.pyplot as plt
```

```python
def plt_plot(y_real):
    fig = plt.figure()
    ax = fig.add_subplot(1, 1, 1)
    ax.plot(range(len(y_real)),y_real, color='b')
    plt.show()
```

```python
source_dire='../../01-TrainingData-additional'
```

```python
file_dire=os.path.join(source_dire,'0'+str(3),'Sensor1')
file_list=os.listdir(file_dire)
mean_list=[]
var_list=[]
for _ in range(1,1+len(file_list)):
    df_temp=pd.read_csv(os.path.join(file_dire,str(_)+".csv"))
    mean_temp=abs(df_temp["vibration_2"]).mean()
    var_temp=abs(df_temp["vibration_2"]).var()
    mean_list.append(mean_temp)
    var_list.append(var_temp)
    print(_,"mean:",mean_temp,"var:",var_temp)

plt_plot(mean_list)
plt_plot(var_list)
```

```python
data_dire=os.path.join(source_dire,'0'+str(1),'Sensor1')
for _ in range(2,10):
    df_signal=pd.read_csv(os.path.join(data_dire,str(_)+".csv"))
    print(os.path.join(data_dire,str(_)+".csv"))
    # plt_plot(df_signal["vibration_1"])
    plt_plot(df_signal["vibration_2"])
    plt_plot(df_signal["vibration_3"])
#     plt_plot(df_signal["current"]) 
    break
```
